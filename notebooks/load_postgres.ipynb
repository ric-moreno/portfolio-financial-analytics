```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()
DB_USER = os.getenv("DB_USER")
DB_PASS = os.getenv("DB_PASS")
DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_NAME = os.getenv("DB_NAME")

engine = create_engine(f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}")

def load_data():
    df_metricas = pd.read_csv("../data/dim_metricas_risco.csv")
    df_retornos_diarios = pd.read_csv("../data/fato_retornos_diarios.csv")
    df_retornos_acum = pd.read_csv("../data/fato_retornos_acumulados.csv")

    df_metricas.columns = df_metricas.columns.str.lower()
    df_retornos_diarios.columns = df_retornos_diarios.columns.str.lower()
    df_retornos_acum.columns = df_retornos_acum.columns.str.lower()

    print(f"Carregando metricas: {len(df_metricas)} linhas")
    df_metricas.to_sql("dim_metricas_risco", engine, schema="financial_analytics", if_exists="replace", index=False)
    print(f"Carregando retornos diarios: {len(df_retornos_diarios)} linhas")
    df_retornos_diarios.to_sql("fato_retornos_diarios", engine, schema="financial_analytics", if_exists="replace", index=False)
    print(f"Carregando retornos acumulados: {len(df_retornos_acum)} linhas")
    df_retornos_acum.to_sql("fato_retornos_acumulados", engine, schema="financial_analytics", if_exists="replace", index=False)
    print("Dados carregados com sucesso no banco de dados!")

if __name__ == "__main__":
    load_data()
```
